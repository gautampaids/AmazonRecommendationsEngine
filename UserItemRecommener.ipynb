# Modeling

```python
train_set, test_set = train_test_split(df, test_size=0.25, random_state=43)
```

```python
user_similarity = pairwise_distances(train_set, metric='cosine')
movie_similarity = pairwise_distances(train_set.T, metric='cosine')
# mean_user_rating = train_set.mean(axis = 1).values.reshape(-1,1)
# ratings_diff = train_set - mean_user_rating
# user_pred = mean_user_rating + user_similarity.dot(ratings_diff)/np.array([np.abs(user_similarity).sum(axis=1)]).T
```

```python
def predict(ratings, similarity, type='user'):
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1).values.reshape(-1,1)
        #You use np.newaxis so that mean_user_rating has same format as ratings
        ratings_diff = ratings - mean_user_rating
        pred = mean_user_rating + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
    elif type == 'movie':
        pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])     
    return pred
```

```python
user_prediction = predict(train_set, user_similarity, type='user')
movie_prediction = predict(train_set, movie_similarity, type='movie')
```

# Evaluation

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(prediction, ground_truth):
    prediction = prediction[ground_truth.nonzero()].flatten() 
    ground_truth = ground_truth[ground_truth.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, ground_truth))
```

```python
print('User-based CF RMSE: ' + str(rmse(user_prediction, test_set.values)))
print('Item-based CF RMSE: ' + str(rmse(movie_prediction.values, test_set.values)))
```

The following measure tells us that User-Item Collborative filtering gives us better results i.e. Users who are similar to you also liked, recommendations based on similar tatest of user
